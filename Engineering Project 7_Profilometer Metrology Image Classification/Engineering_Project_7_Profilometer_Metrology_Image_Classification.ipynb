```python
import os, random, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix
import tensorflow as tf

# ----------------------------
# 0) Reproducibility
# ----------------------------
SEED = 7
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ----------------------------
# 1) Paths / settings
# ----------------------------
DATA_DIR = "/content/project8_data"
OUT_DIR  = "/content/project8_output"
os.makedirs(DATA_DIR, exist_ok=True)
os.makedirs(OUT_DIR, exist_ok=True)

IMG_SIZE = 128
N_IMAGES_TOTAL = 1000

CLASSES = ["normal", "spike", "tilt_drift", "scratch_gouge", "vibration_ripple"]
N_CLASSES = len(CLASSES)
N_PER_CLASS = N_IMAGES_TOTAL // N_CLASSES  # 200 each (exactly 1000 total)

# Split ratios
TRAIN_FRAC = 0.70
VAL_FRAC   = 0.15
TEST_FRAC  = 0.15

BATCH_SIZE = 32
EPOCHS = 35  # early stopping will usually stop earlier

# ----------------------------
# 2) Synthetic profilometer image generator
#    (Each image is a 2D "height map" rendered to grayscale.)
# ----------------------------
def normalize_to_uint8(z: np.ndarray) -> np.ndarray:
    z = z.astype(np.float32)
    z = z - z.min()
    z = z / (z.max() + 1e-8)
    return (255.0 * z).clip(0, 255).astype(np.uint8)

def smooth2d(z: np.ndarray, k: int = 5) -> np.ndarray:
    """Simple box blur without external deps."""
    if k <= 1:
        return z
    pad = k // 2

    # Pad the input image z to handle borders during convolution
    zp = np.pad(z, ((pad, pad), (pad, pad)), mode="reflect")

    # Calculate the integral image of the padded array.
    # Add an extra row and column of zeros for simpler boundary handling in sum calculation.
    H_zp, W_zp = zp.shape
    ii = np.zeros((H_zp + 1, W_zp + 1), dtype=np.float32)
    ii[1:, 1:] = np.cumsum(np.cumsum(zp, axis=0), axis=1)

    out = np.zeros_like(z, dtype=np.float32)
    H, W = z.shape # Original dimensions

    # For each pixel (i, j) in the original image (and output),
    # determine the k x k window in the padded image (zp) to sum.
    # The window starts at zp[i, j] and ends at zp[i + k - 1, j + k - 1].
    for i in range(H):
        for j in range(W):
            r_start = i
            c_start = j
            r_end = i + k - 1
            c_end = j + k - 1

            # Use the integral image to sum the k x k window
            # ii[r_idx, c_idx] represents sum of zp[0..r_idx-1, 0..c_idx-1]
            s = ii[r_end + 1, c_end + 1] \
              - ii[r_start, c_end + 1] \
              - ii[r_end + 1, c_start] \
              + ii[r_start, c_start]

            out[i, j] = s / (k * k)
    return out

def base_surface(img_size=128) -> np.ndarray:
    """Base wafer/metrology surface: smooth low-frequency texture + small noise."""
    z = np.random.normal(0, 0.25, (img_size, img_size)).astype(np.float32)
    z = smooth2d(z, k=7)
    z += np.random.normal(0, 0.03, z.shape).astype(np.float32)
    return z

def add_spikes(z: np.ndarray, n_spikes=None) -> np.ndarray:
    """Add bright gaussian-like spikes (particles / stylus artifacts)."""
    H, W = z.shape
    if n_spikes is None:
        n_spikes = np.random.randint(1, 4)
    for _ in range(n_spikes):
        cx = np.random.randint(10, W-10)
        cy = np.random.randint(10, H-10)
        amp = np.random.uniform(0.8, 2.0)
        sigma = np.random.uniform(1.2, 3.0)
        yy, xx = np.mgrid[0:H, 0:W]
        z += amp * np.exp(-((xx-cx)**2 + (yy-cy)**2) / (2*sigma**2)).astype(np.float32)
    return z

def add_tilt_drift(z: np.ndarray) -> np.ndarray:
    """Add a global gradient + mild curvature (tool drift / wafer tilt)."""
    H, W = z.shape
    ax = np.random.uniform(-1.2, 1.2)
    ay = np.random.uniform(-1.2, 1.2)
    yy, xx = np.mgrid[0:H, 0:W]
    xxn = (xx - W/2) / (W/2)
    yyn = (yy - H/2) / (H/2)
    z += 0.7 * (ax * xxn + ay * yyn).astype(np.float32)
    # mild bow/curvature
    curv = np.random.uniform(-0.4, 0.4)
    z += (curv * (xxn**2 + yyn**2)).astype(np.float32)
    return z

def add_scratch_gouge(z: np.ndarray) -> np.ndarray:
    """Add a line-like defect (scratch/trench) with jitter."""
    H, W = z.shape
    yy, xx = np.mgrid[0:H, 0:W]

    # random line angle and position
    angle = np.random.choice([0, 20, 45, 70, 90, 110, 135, 160]) * np.pi / 180.0
    x0 = np.random.uniform(0.2*W, 0.8*W)
    y0 = np.random.uniform(0.2*H, 0.8*H)

    # line equation: distance from line
    # normal vector (nx, ny)
    nx, ny = np.cos(angle), np.sin(angle)
    dist = (xx - x0) * (-ny) + (yy - y0) * (nx)  # signed distance

    width = np.random.uniform(1.5, 4.0)
    depth = np.random.uniform(0.7, 1.7)  # gouge depth
    # gouge = negative ridge along line
    z -= depth * np.exp(-(dist**2) / (2*width**2)).astype(np.float32)

    # add some raggedness along scratch direction
    rag = np.random.normal(0, 0.05, z.shape).astype(np.float32)
    z += rag * np.exp(-(dist**2) / (2*(width*1.5)**2)).astype(np.float32)

    return z

def add_vibration_ripple(z: np.ndarray) -> np.ndarray:
    """Add periodic ripple patterns (mechanical vibration / periodic noise)."""
    H, W = z.shape
    yy, xx = np.mgrid[0:H, 0:W]

    # random frequency and direction
    theta = np.random.uniform(0, np.pi)
    fx = np.cos(theta)
    fy = np.sin(theta)
    freq = np.random.uniform(4.0, 12.0)  # cycles across image
    phase = np.random.uniform(0, 2*np.pi)
    amp = np.random.uniform(0.35, 0.9)

    # normalized coordinate along direction
    u = (xx * fx + yy * fy) / max(H, W)
    ripple = amp * np.sin(2*np.pi*freq*u + phase).astype(np.float32)

    # add slight second harmonic
    ripple += 0.25*amp * np.sin(2*np.pi*(freq*2)*u + phase*0.7).astype(np.float32)

    z += ripple
    return z

def make_image_for_class(class_name: str, img_size: int = 128) -> np.ndarray:
    z = base_surface(img_size)

    if class_name == "normal":
        # keep mostly smooth
        z += np.random.normal(0, 0.02, z.shape).astype(np.float32)

    elif class_name == "spike":
        z = add_spikes(z)

    elif class_name == "tilt_drift":
        z = add_tilt_drift(z)

    elif class_name == "scratch_gouge":
        z = add_scratch_gouge(z)

    elif class_name == "vibration_ripple":
        z = add_vibration_ripple(z)

    else:
        raise ValueError(f"Unknown class: {class_name}")

    # final light smoothing to mimic measurement filtering
    if class_name in ["normal", "tilt_drift"]:
        z = smooth2d(z, k=3)

    # normalize to image
    img = normalize_to_uint8(z)
    return img  # shape (H,W) uint8

def write_dataset(root_dir: str):
    # Create split dirs
    for split in ["train", "val", "test"]:
        for c in CLASSES:
            os.makedirs(os.path.join(root_dir, split, c), exist_ok=True)

    for c in CLASSES:
        n = N_PER_CLASS
        idx = np.arange(n)
        rng = np.random.default_rng(SEED + hash(c) % 1000)
        rng.shuffle(idx)

        n_train = int(TRAIN_FRAC * n)
        n_val   = int(VAL_FRAC * n)
        train_idx = idx[:n_train]
        val_idx   = idx[n_train:n_train+n_val]
        test_idx  = idx[n_train+n_val:]

        for k_idx, ii in enumerate(train_idx):
            img = make_image_for_class(c, IMG_SIZE)
            path = os.path.join(root_dir, "train", c, f"{c}_{ii:04d}.png")
            plt.imsave(path, img, cmap="gray", vmin=0, vmax=255)

        for k_idx, ii in enumerate(val_idx):
            img = make_image_for_class(c, IMG_SIZE)
            path = os.path.join(root_dir, "val", c, f"{c}_{ii:04d}.png")
            plt.imsave(path, img, cmap="gray", vmin=0, vmax=255)

        for k_idx, ii in enumerate(test_idx):
            img = make_image_for_class(c, IMG_SIZE)
            path = os.path.join(root_dir, "test", c, f"{c}_{ii:04d}.png")
            plt.imsave(path, img, cmap="gray", vmin=0, vmax=255)

    print("Dataset created at:", root_dir)
    print("Total images:", N_IMAGES_TOTAL, f"({N_PER_CLASS} per class)")

# Generate dataset (always; delete folder if you want regeneration)
write_dataset(DATA_DIR)

# Quick peek at one image per class
plt.figure(figsize=(12, 2.4))
for i, c in enumerate(CLASSES):
    sample_path = next(iter(tf.io.gfile.glob(os.path.join(DATA_DIR, "train", c, "*.png"))))
    img = plt.imread(sample_path)
    plt.subplot(1, len(CLASSES), i+1)
    plt.imshow(img, cmap="gray")
    plt.title(c, fontsize=9)
    plt.axis("off")
plt.suptitle("Sample synthetic profilometer images")
plt.show()

# ----------------------------
# 3) Load data (tf.data)
# ----------------------------
def make_ds(split: str, shuffle: bool):
    ds = tf.keras.utils.image_dataset_from_directory(
        os.path.join(DATA_DIR, split),
        labels="inferred",
        label_mode="int",
        image_size=(IMG_SIZE, IMG_SIZE),
        color_mode="grayscale",
        batch_size=BATCH_SIZE,
        shuffle=shuffle,
        seed=SEED,
    )
    return ds

train_ds = make_ds("train", shuffle=True)
val_ds   = make_ds("val", shuffle=False)
test_ds  = make_ds("test", shuffle=False)

class_names = train_ds.class_names
print("Class names:", class_names)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)
test_ds  = test_ds.cache().prefetch(AUTOTUNE)

# ----------------------------
# 4) Model: CNN + augmentation + regularization (avoid overfitting)
# ----------------------------
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.08),
    tf.keras.layers.RandomZoom(0.08),
    tf.keras.layers.RandomContrast(0.12),
], name="augment")

weight_decay = 1e-4

def conv_block(x, filters, k=3, drop=0.0):
    x = tf.keras.layers.Conv2D(filters, k, padding="same",
                               kernel_regularizer=tf.keras.regularizers.l2(weight_decay))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    x = tf.keras.layers.Conv2D(filters, k, padding="same",
                               kernel_regularizer=tf.keras.regularizers.l2(weight_decay))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    x = tf.keras.layers.MaxPool2D()(x)
    if drop > 0:
        x = tf.keras.layers.Dropout(drop)(x)
    return x

inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 1))
x = data_augmentation(inputs)
x = tf.keras.layers.Rescaling(1./255)(x)

x = conv_block(x, 32, drop=0.10)
x = conv_block(x, 64, drop=0.15)
x = conv_block(x, 128, drop=0.20)

x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(128, activation="relu",
                          kernel_regularizer=tf.keras.regularizers.l2(weight_decay))(x)
x = tf.keras.layers.Dropout(0.35)(x)
outputs = tf.keras.layers.Dense(N_CLASSES, activation="softmax")(x)

model = tf.keras.Model(inputs, outputs)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=6, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-5),
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# Save model
model_path = os.path.join(OUT_DIR, "profilometer_defect_cnn.keras")
model.save(model_path)
print("Saved model:", model_path)

# ----------------------------
# 5) Evaluate: confusion matrix + per-class accuracy table
# ----------------------------
# Gather predictions on test set
y_true = []
y_pred = []

for batch_imgs, batch_labels in test_ds:
    probs = model.predict(batch_imgs, verbose=0)
    preds = np.argmax(probs, axis=1)
    y_true.extend(batch_labels.numpy().tolist())
    y_pred.extend(preds.tolist())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

cm = confusion_matrix(y_true, y_pred, labels=np.arange(N_CLASSES))

print("\nConfusion Matrix (rows=true, cols=pred):")
print(pd.DataFrame(cm, index=class_names, columns=class_names))

# Per-class accuracy = correct / total true for that class
row_sums = cm.sum(axis=1).clip(min=1)
per_class_acc = np.diag(cm) / row_sums

acc_table = pd.DataFrame({
    "class": class_names,
    "support_true_count": row_sums,
    "per_class_accuracy": per_class_acc
}).sort_values("per_class_accuracy", ascending=False)

print("\nPer-class accuracy table:")
print(acc_table.to_string(index=False))

# Save table
acc_csv = os.path.join(OUT_DIR, "per_class_accuracy.csv")
acc_table.to_csv(acc_csv, index=False)
print("Saved per-class accuracy:", acc_csv)

# Plot training curves
plt.figure(figsize=(7, 4))
plt.plot(history.history["accuracy"], label="train_acc")
plt.plot(history.history["val_accuracy"], label="val_acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Curves")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# ----------------------------
# 6) Demo: predict labels for 10 unseen test images (and show them)
# ----------------------------
# Collect file paths from test directory
test_files = []
test_labels = []
for ci, cname in enumerate(class_names):
    files = tf.io.gfile.glob(os.path.join(DATA_DIR, "test", cname, "*.png"))
    for f_item in files:
        test_files.append(f_item)
        test_labels.append(ci)

# pick 10 random
idxs = np.random.default_rng(SEED).choice(len(test_files), size=10, replace=False)

plt.figure(figsize=(14, 6))
for i, idx in enumerate(idxs, start=1):
    f_item = test_files[idx]
    true_label = class_names[test_labels[idx]]

    # Load + preprocess
    img = tf.keras.utils.load_img(f_item, color_mode="grayscale", target_size=(IMG_SIZE, IMG_SIZE))
    arr = tf.keras.utils.img_to_array(img)  # (H,W,1)
    arr_b = np.expand_dims(arr, axis=0)

    prob = model.predict(arr_b, verbose=0)[0]
    pred_idx = int(np.argmax(prob))
    pred_label = class_names[pred_idx]
    conf = float(np.max(prob))

    plt.subplot(2, 5, i)
    plt.imshow(arr.squeeze(), cmap="gray")
    plt.axis("off")
    plt.title(f"T:{true_label}\nP:{pred_label} ({conf:.2f})", fontsize=9)

plt.suptitle("Demo: 10 unseen test images  True vs Predicted", fontsize=12)
plt.show()

print("\nOutputs saved in:", OUT_DIR)
print(" - profilometer_defect_cnn.keras")
print(" - per_class_accuracy.csv")
print("\nIn Colab: left sidebar  Files  project8_output  download artifacts")
```
